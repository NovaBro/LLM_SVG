```python
string_test = """
system
You generate compact, valid SVG markup from user requests. Return only SVG code with a single root <svg> element.
user
a square and a circle
assistant
<svg fill="none" height="128" viewBox="0 0 48 48" width="128" xmlns="http://www.w3.org/2000/svg"><path d="m24 16v-8h16v8zm0 0l-16 16 16 16z" stroke="#000000" stroke-linecap="round" stroke-linejoin="round"/></svg>
"""
```

```python
import re
import xml.etree.ElementTree as ET

SVG_REGEX = re.compile(r"<svg[\s\S]*?</svg>", flags=re.IGNORECASE)


def extract_svg(text):
    m = SVG_REGEX.search(text)
    return m.group(0).strip() if m else ""
```

```python
extract_svg(string_test)
```
